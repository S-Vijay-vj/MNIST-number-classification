```python
model = tf.keras.models.load_model('mnist_text_classfier.model')
```

```python
def image_classification(img_dir):
    img = tf.keras.utils.load_img(img_dir,target_size=(28,28),color_mode='grayscale')
    img_array=tf.keras.utils.img_to_array(img)
    img_array = tf.keras.utils.normalize(img_array,axis=1)
    reshaped_array = np.array([img_array])
    ans = model.predict(reshaped_array)
    classification = np.argmax(ans)
    print (f'The given image is number - {classification}')
    return plt.imshow(img_array)
    
```
